# Regression

```python
data=pd.read_csv("classification.csv").dropna().iloc[:,:]
```

```python
# ETL
data=pd.read_csv("classification.csv").dropna()

# Select variables & data train & data new

features_x=['alpha', 'delta', 'u', 'g', 'r', 'i', 'z']
features_y=['class']
X=data.loc[:,features_x]
y=data.loc[:,features_y]
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
y=le.fit_transform(np.array(y))

# Split
X_train, X_test, y_train, y_test=split_test_train(X,y,test_size=0.2)

# Scaler
direccion='models/'
X_train_scaler=Scaler(df=X_train,direction=direccion)
with open('models/model_scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)
X_test_scaler=scaler.transform(X_test)
```

```python
model, pred = algorithm_pipeline(X_train_data=X_train_scaler, X_test_data=X_test_scaler,y_train_data=y_train,y_test_data=y_test, model=XGBClassifier(), cv=5)
```

```python
print(model.best_score_)
print(model.best_params_)
```

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict(X_test_scaler)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
print("/n")
y_pred = model.predict(X_train_scaler)
accuracy = accuracy_score(y_train, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
