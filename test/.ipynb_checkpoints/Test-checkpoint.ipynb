# LSTM Autoencoder Anomaly Detection

```python
# ETL
data=pd.read_csv("lstm.csv").dropna()
# Select variables & data train & data new
porcentaje_train=0.9
direccion=''

data_select_dataframe=data.iloc[:int(data.shape[0]*porcentaje_train),:]
data_select=np.array(data.loc[:,['Open','High','Low']].iloc[:int(data.shape[0]*porcentaje_train),:])
data_select_new=np.array(data.loc[:,['Open','High','Low']].iloc[int(data.shape[0]*porcentaje_train):,:])
data_select_new_dataframe=data.iloc[int(data.shape[0]*porcentaje_train):,:]
data_select=Scaler(df=data_select,direction=direccion)
```

```python
# Create model
# Create Arrays
t_stp=10
data_array=create_sequences(X=data_select,time_steps=t_stp)
# model
dim_out=128
dim_in=64
dp=0.3
patience=5
direccion=''
bs=int(data_array.shape[0]/5)
n_epochs=1
verbose=1
model=model_lstm(data=data_array,time_steps=t_stp,dim_out=dim_out,dim_in=dim_in,dp=dp)
# Training
training(data=data_array,model=model,patience=patience,direccion=direccion,bs=bs,n_epochs=n_epochs,verbose=verbose)
```

```python
porcentaje_max=0.05
direction=''
resta_global_total=P_value_calculate(model_dir='model_lstm.h5',data_array=data_array,porcentaje_max=porcentaje_max,direction=direction)
```

```python
value_limited=np.load(direction+'p_value.npy')
num_max=9
direccion=''
result=Export_results_xlsx(resta_global_total=resta_global_total,value_limited=value_limited,data_select_dataframe=data_select_dataframe,num_max=num_max,t_stp=t_stp,direccion=direccion)
```

```python
direccion=''
for prop in result.columns[1:-1]:
    fig,ax=plt.subplots(figsize=(20, 16)) 
    plt.plot(result['index'],result[prop],marker="o",label=prop)
    plt.plot(result[result['Out/In']==1]['index'],result[result['Out/In']==1][prop],marker="o",color="red",linestyle='')
    plt.title(prop, fontdict=None, loc='center', pad=None,fontsize=30)
    plt.savefig(direccion+'Grap_'+prop+'.png', bbox_inches='tight', pad_inches=0.0)
```

```python
# Cargamos el modelo y hacemos inferencia
def P_value_calculate(model,data_array,porcentaje_max,direction)
m="model_lstm.h5"
model=keras.models.load_model(m)
data_array_pred =model.predict(data_array)
resta=np.abs(data_array_pred - data_array)
resta_global_total=resta.mean(axis=2).mean(axis=1)
```

```python
# P_Value
porcentaje_max=0.005
value_limited=sorted(resta_global_total,reverse=True)[int(len(resta_global_total)*porcentaje_max)]
# Save p_Value
with open('p_value.npy', 'wb') as f:
    np.save(f, np.array(value_limited))
```

```python

# Calculate id (del array) out/in
resta_global_binary=[]
for r in resta_global_total:
    if r>value_limited:
        resta_global_binary.append(1)
    else:
        resta_global_binary.append(0)
        
# Calculate los id que son out/in reales
out_list=[]
for i in range(0,len(resta_global_binary)):
    if resta_global_binary[i]==1:
        list_pre=list(range(i,i+t_stp))
        for l in list_pre:
                out_list.append(l)
    else:
        continue

dictionary_counts=Counter(out_list)
dictionary_keys=list(Counter(out_list).keys())
final=[]
for key in dictionary_keys:
    if dictionary_counts[key]>9:
        final.append(key) 

# Merge results out/in
Final=pd.DataFrame(final)
Final['Out/In']=1
Final.columns=['Index','Out/In']
Final.set_index('Index', inplace=True)
result = pd.merge(data_select_dataframe, Final, how="left",left_index=True, right_index=True).fillna(0)
result.to_excel("resultados.xlsx")
result=result.reset_index()
```

```python
# Visualizacion

for prop in result.columns[1:-1]:
    fig,ax=plt.subplots(figsize=(20, 16)) 
    plt.plot(result['index'],result[prop],marker="o",label=prop)
    plt.plot(result[result['Out/In']==1]['index'],result[result['Out/In']==1][prop],marker="o",color="red",linestyle='')
    plt.title(prop, fontdict=None, loc='center', pad=None,fontsize=30)
    plt.savefig('Produccion/Results/Grap/'+prop+'.png', bbox_inches='tight', pad_inches=0.0)
```
